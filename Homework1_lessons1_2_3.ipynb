```python
import numpy as np
def length(vector):
    sum_sqr = 0
    for i in range(len(vector)):
        sum_sqr += vector[i]**2
    return np.sqrt(sum_sqr)
```

```python
import matplotlib.pyplot as plt
import math
```

```python
x = []
x0 = 1
y0 = 1
y1 = []
y2 = []
r = 2
for i in range(-r * 10000, r * 10000):
    x_= i / 10000 + x0
    x.append(x_)
    y1.append(math.sqrt(r**2 - (x_ - x0)**2) + y0)
    y2.append(- math.sqrt(r**2 - (x_ - x0)**2) + y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Окружность')
plt.gca().set_aspect('equal', adjustable='box')
```

<b>3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.</b>

```python
import math
# определим угол на промежутке от 0 до  двух Пи
# будем возвращать угол в радианах
def desc_to_polar(x, y):
    r = math.sqrt(x ** 2 + y ** 2)
    if x == 0:
        if y > 0:
            return 0.5 * np.pi, r
        elif y < 0:
            return 1.5 * np.pi, r
        else:
            print('x = 0 и y = 0, r и угол не определены')
            
    if x < 0:
        return np.arctan(y / x) + np.pi, r
    
    if x > 0:
        if y >= 0:
            return np.arctan(y / x), r
        else:
            return np.arctan(y / x) + 2 * np.pi, r
```

<b>4. Задание (в программе)
Решите систему уравнений:<br>
y = x**2 – 1<br>
exp(x) + x∙(1 – y) = 1</b>

Сначала посмотрим на количество корней у системы уравнений и примерные координаты решений, потом будем вблизи искать точные численные решения:

```python
x = np.arange(-5, 5, 0.1)
y1 = x ** 2 - 1 
y2 = []
for x_ in x:
    y2.append((math.exp(x_) + x_ - 1) / x_)
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(labels=['y = x ** 2 - 1', '(exp(x) + x - 1) / x'])
plt.grid()
```

```python
from scipy.optimize import fsolve
import math

def equations(p):
    x, y = p
    return (x ** 2 - 1 - y, math.exp(x) + x *(1 - y) - 1)

x, y =  fsolve(equations, (-1, 1))

equations((x, y))
```

```python
import scipy.optimize as opt
from numpy import exp

def f(variables) :
    (x,y) = variables
    first_eq = x ** 2 - 1 - y
    second_eq = exp(x) + x *(1 - y) - 1
    return [first_eq, second_eq]
solutions = []
solutions.append(opt.fsolve(f, (-1.5, 1.8) ))
solutions.append(opt.fsolve(f, (2.5, 6) ))
solutions.append(opt.fsolve(f, (4.1, 16) ))
# массив с численными решениями системы уравнений: 
solutions
```

```python
x = np.arange(-5, 5, 0.1)
y1 = x ** 2 - 1 
y2 = []
for x_ in x:
    y2.append((math.exp(x_) + x_ - 1) / x_)
plt.plot(x, y1, label='y = x ** 2 - 1')
plt.plot(x, y2)
for i in range(len(solutions)):
    plt.scatter(solutions[i][0], solutions[i][1], color='orange', s=40, marker='o')
plt.grid()
```
