1. Задание 2) Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
def length(vector):
    sum_sqr = 0
    for i in range(len(vector)):
        sum_sqr += vector[i]**2
    return np.sqrt(sum_sqr)
```

```python
a = np.array([3, 4])
length(a)
```

3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
import math
```

```python
x = []
x0 = 1
y0 = 1
y1 = []
y2 = []
r = 2
for i in range(-r * 10000, r * 10000):
    x_= i / 10000 + x0
    x.append(x_)
    y1.append(math.sqrt(r**2 - (x_ - x0)**2) + y0)
    y2.append(- math.sqrt(r**2 - (x_ - x0)**2) + y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Окружность')
plt.gca().set_aspect('equal', adjustable='box')
```

```python
x = []
x0 = 2
y0 = 2
b = 2
a = 4
y1 = []
y2 = []
for i in range(10000 * (- a), 10000 * a):
    x_= i / 10000 + x0
    x.append(x_)
    y1.append(math.sqrt((1 - ((x_ - x0)**2) / a**2) * b**2) + y0)
    y2.append(- math.sqrt((1 - ((x_ - x0)**2) / a**2) * b**2) + y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Эллипс')
plt.gca().set_aspect('equal', adjustable='box')
```

```python
x_left = []
x_right = []
x0 = 2
y0 = 2
b = 2
a = 4
y1 = []
y2 = []
y3 = []
y4 = []
for i in range(10000 * (- 2 * a), 10000 * (- a)):
    x_= i / 10000 + x0
    x_left.append(x_)
    y1.append(math.sqrt(((x_ - x0)**2)/ a**2 - 1) * b**2 + y0)
    y2.append(-math.sqrt(((x_ - x0)**2)/ a**2 - 1) * b**2 + y0)
for i in range(10000 * a, 10000 * 2 * a):
    x_= i / 10000 + x0
    x_right.append(x_)
    y3.append(math.sqrt(((x_ - x0)**2)/ a**2 - 1) * b**2 + y0)
    y4.append(- math.sqrt(((x_ - x0)**2)/ a**2 - 1) * b**2 + y0)

plt.plot(x_left, y1)
plt.plot(x_left, y2)
plt.plot(x_right, y3)
plt.plot(x_right, y4)
plt.title('Гипербола')
plt.xlabel('x')
plt.ylabel('y')
```

5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 5*Y
Z_paralell = 2*X + 5*Y + 20
ax.plot_wireframe(X, Y, Z, color='red')
ax.plot_wireframe(X, Y, Z_paralell)
show()
```

Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
ax.set_aspect("equal")
X = np.arange(-2, 2, 0.2)
Y = np.arange(-2, 2, 0.2)
X, Y = np.meshgrid(X, Y)
Z1 = np.sqrt(X**2 + Y**2)
Z2 = - np.sqrt(X**2 + Y**2)
ax.plot_surface(X, Y, Z1, color='green')
ax.plot_surface(X, Y, Z2, color='green')
plt.title('Пирамида')
show()
```
