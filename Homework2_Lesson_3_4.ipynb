Напишите код, который будет переводить полярные координаты в декартовы.

```python
import math
# определим угол на промежутке от 0 до  двух Пи
# будем возвращать угол в радианах
def desc_to_polar(x, y):
    r = math.sqrt(x ** 2 + y ** 2)
    if x == 0:
        if y > 0:
            return 0.5 * np.pi, r
        elif y < 0:
            return 1.5 * np.pi, r
        else:
            print('x = 0 и y = 0, r и угол не определены')
            
    if x < 0:
        return np.arctan(y / x) + np.pi, r
    
    if x > 0:
        if y >= 0:
            return np.arctan(y / x), r
        else:
            return np.arctan(y / x) + 2 * np.pi, r
```

1. Напишите код, проверяющий любую из теорем сложения и умножения вероятности на примере рулетки. 

```python
# Если рулетка исправна, то попадание мячика на каждый номер -- независимое событие. 
# Значит, вероятность выпадения, например, P(числа 1 и числа 26) = P( выпадения числа 1) + P(выпадения числа 26)
# P выпадения числа, например, 1, мы можем посчитать как количество выпадения 1 поделить на количество проведенных игр, 
# А P(выпадения 1 и выпадения 26) = количества выпадения 1 и 26 разделить на количество проведенных игр 

from collections import Counter
# количество ячеек у рулетки:
N = 37
# количество игр
count_games = 1000
# пустой список под подсчет выпавших номеров
lucky_numbers = []
# будем проверять теорему сложения вероятности на двух событиях, на двух выпавших номерах, например
A = 0
B = 36
# генерируем массив со значениями выпавших чисел
for i in range(0, count_games):
    lucky_number = np.random.randint(0, 37)
    lucky_numbers.append(lucky_number)
count_lucky_numbers = Counter(lucky_numbers)
collect_probab = {}
for lucky_number, quantity in count_lucky_numbers.items():
    # ищем вероятность выпадания для каждого числа из набора рулетки
    probability = quantity / count_games
    # получаем словарь со значениями возможных номеров и их вероятностей
    collect_probab[lucky_number] = probability
# проверка, сумма всех вероятностей должна быть равна 1
summ = 0
for key, item in collect_probab.items():
    summ += item
print(f'Проверка, сумма всех вероятностей {summ}')
print(f'Вероятность события выпадения {A} и события {B} равна:')
print(f'{(count_lucky_numbers[A] + count_lucky_numbers[B]) / count_games}') 
print(f'Вероятность события выпадения {A} равна:')
print(collect_probab[A])
print(f'Вероятность события выпадения {B} равна:')
print(collect_probab[B]) 
print(f'Сумма вероятностей выпадения {A} и {B} равна: {collect_probab[A] + collect_probab[B]}')
```
